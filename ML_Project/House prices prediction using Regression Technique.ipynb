### Data Preprocessing

```python
X = df.drop(["price", "id", "date"], axis=1)

y = df["price"]
```

```python
data_corr = X.corr()
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(data_corr, cbar = True,  square = True, annot=True, fmt= '.2f',annot_kws={'size': 15},cmap= 'coolwarm')
plt.title('Correlation between features');
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.2, random_state=21)
```

### Evaluating the Algorithms

### Linear Regression

```python
from sklearn.linear_model import LinearRegression

model_1 = LinearRegression()

model_1.fit(X_train,y_train)
```

```python
y_pred = model_1.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error , mean_squared_error , r2_score
```

```python
# finding the r2 score or the variance
r2 = r2_score(y_test, y_pred)
print("R2 Score:", r2)
```

```python
print("MAE: " , mean_absolute_error( y_test, y_pred)) 
print("MSE: " ,  mean_squared_error(y_test,y_pred)) 
print("RMSE: " , np.sqrt(mean_squared_error(y_test, y_pred)))
```

## Gradient Boostings

```python
from sklearn.ensemble import GradientBoostingRegressor

from math import *

model_2 = GradientBoostingRegressor(n_estimators = 100, max_depth = 5, min_samples_split = 2, learning_rate = 0.1)
model_2.fit(X_train, y_train)

y_pred = model_2.predict(X_test)


# finding the mean_squared error
mse = mean_squared_error(y_test, y_pred)
print("RMSE Error:", np.sqrt(mse))

# finding the r2 score or the variance
r2 = r2_score(y_test, y_pred)
print("R2 Score:", r2)
```

```python
feat_importances = pd.Series(model_2.feature_importances_, index= X.columns)

feat_importances.nlargest(25).plot(kind='barh',figsize=(10,10))
```

Gradient Boosting displayed the best performance for this Dataset and can be used for deploying purposes
